```python
import pandas as pd

# Criando um DataFrame de exemplo
data = {'Cidade': ['Nova York', 'Los Angeles', 'Chicago', 'Houston', 'Miami'],
        'População': [8405837, 3990456, 2716000, 2327886, 413201]
       }

df = pd.DataFrame(data)

# Criando um dicionário com os valores-chave e seus respectivos valores para a nova coluna
estado_dict = {
    'Nova York': 'NY',
    'Los Angeles': 'CA',
    'Chicago': 'IL',
    'Houston': 'TX',
    'Miami': 'FL'
}

# Usando o método .map() para criar uma nova coluna 'Estado' com base no mapeamento do dicionário
df['Estado'] = df['Cidade'].map(estado_dict)

# Exibindo o DataFrame resultante
print(df)
```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores_serie1 = [10, 15, 7, 12]
valores_serie2 = [5, 8, 9, 14]

# Crie uma figura e um eixo
fig, ax = plt.subplots()

# Plote as séries de dados com barras
bar_width = 0.35
index = range(len(categorias))
bar1 = ax.bar(index, valores_serie1, bar_width, label='Série 1')
bar2 = ax.bar([i + bar_width for i in index], valores_serie2, bar_width, label='Série 2')

# Adicione rótulos aos eixos
ax.set_xlabel('Categorias')
ax.set_ylabel('Valores')

# Adicione rótulos para as categorias no eixo x
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(categorias)

# Adicione uma legenda ao gráfico
ax.legend()

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Dados de exemplo
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [10, 15, 7, 12]

# Cores correspondentes às categorias
cores = ['blue', 'green', 'red', 'purple']

# Crie uma figura e um eixo
fig, ax = plt.subplots()

# Plote as barras com cores personalizadas
bars = ax.bar(categorias, valores, color=cores)

# Crie uma lista de patches coloridos para a legenda
legend_patches = [Patch(color=color, label=categoria) for categoria, color in zip(categorias, cores)]

# Adicione a legenda personalizada ao gráfico
ax.legend(handles=legend_patches, title='Categorias')

# Adicione rótulos aos eixos
ax.set_xlabel('Categorias')
ax.set_ylabel('Valores')

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Seu dicionário de mapeamento de categorias para etiquetas
mapeamento = {
    'Categoria1': 'Etiqueta 1',
    'Categoria2': 'Etiqueta 2',
    'Categoria3': 'Etiqueta 3'
}

# Dados de exemplo para o gráfico de barras
categorias = ['Categoria1', 'Categoria2', 'Categoria3']
valores = [10, 20, 15]

# Crie o gráfico de barras
plt.bar(categorias, valores)

# Crie uma lista de handles e etiquetas para a legenda
handles = [plt.Rectangle((0,0),1,1,fc="white", edgecolor="black", lw=2) for _ in categorias]
etiquetas = [mapeamento[categoria] for categoria in categorias]

# Adicione a legenda ao gráfico
plt.legend(handles, etiquetas, title='Categorias')

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Seu dicionário de mapeamento de categorias para etiquetas e cores
mapeamento = {
    'Categoria1': {'etiqueta': 'Etiqueta 1', 'cor': 'red'},
    'Categoria2': {'etiqueta': 'Etiqueta 2', 'cor': 'blue'},
    'Categoria3': {'etiqueta': 'Etiqueta 3', 'cor': 'green'}
}

# Dados de exemplo para o gráfico de barras
categorias = ['Categoria1', 'Categoria2', 'Categoria3']
valores = [10, 20, 15]

# Crie o gráfico de barras
bars = plt.bar(categorias, valores)

# Crie uma lista de handles e etiquetas para a legenda
handles = [Patch(color=mapeamento[categoria]['cor']) for categoria in categorias]
etiquetas = [mapeamento[categoria]['etiqueta'] for categoria in categorias]

# Adicione a legenda ao gráfico
plt.legend(handles, etiquetas, title='Categorias')

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crie uma figura com 3 subplots em uma única linha
plt.figure(figsize=(5, 4))  # Defina o tamanho da figura

# Primeiro subplot
plt.subplot(131)  # 1 linha, 3 colunas, primeiro subplot
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
plt.plot(x1, y1)
plt.title('Gráfico 1')

# Segundo subplot
plt.subplot(132)  # 1 linha, 3 colunas, segundo subplot
x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)
plt.plot(x2, y2)
plt.title('Gráfico 2')

# Terceiro subplot
plt.subplot(133)  # 1 linha, 3 colunas, terceiro subplot
x3 = np.linspace(0, 10, 100)
y3 = x3
plt.plot(x3, y3)
plt.title('Gráfico 3')

# Ajuste o layout dos subplots
plt.tight_layout()

# Exiba a figura
plt.show()
```
