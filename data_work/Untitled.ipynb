```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo para os histogramas
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)
data3 = np.random.randn(1000)
data4 = np.random.randn(1000)

# Crie uma figura com uma matriz de subplots 2x2 (4 facet)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Crie os histogramas em cada subplot
axs[0, 0].hist(data1, bins=20, color='blue', alpha=0.7)
axs[0, 0].set_title('Facet 1')

axs[0, 1].hist(data2, bins=20, color='green', alpha=0.7)
axs[0, 1].set_title('Facet 2')

axs[1, 0].hist(data3, bins=20, color='red', alpha=0.7)
axs[1, 0].set_title('Facet 3')

axs[1, 1].hist(data4, bins=20, color='purple', alpha=0.7)
axs[1, 1].set_title('Facet 4')

# Adicione um título global para a figura
plt.suptitle('Histogramas com 4 Facets', fontsize=16)

# Ajuste automaticamente o layout para evitar sobreposição
plt.tight_layout()

# Exiba a figura
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponhamos que você tenha um DataFrame com uma coluna chamada "Sexo" contendo valores "M" e "F"
# Aqui está um exemplo de como criar um DataFrame fictício para demonstração:
data = {'Sexo': ["M", "F", "M", "F", "M", "M", "F", "F", "M", "M"]}
df = pd.DataFrame(data)

# Conte as ocorrências de cada categoria
contagem = df['Sexo'].value_counts()

# Crie o gráfico de barras horizontais
plt.barh(contagem.index, contagem.values)

# Adicione rótulos aos eixos
plt.xlabel('Contagem')
plt.ylabel('Sexo')

# Exiba o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponhamos que você tenha um DataFrame com uma coluna chamada "Sexo" contendo valores "M" e "F"
# Aqui está um exemplo de como criar um DataFrame fictício para demonstração:
data = {'Sexo': ["M", "F", "M", "F", "M", "M", "F", "F", "M", "M"]}
df = pd.DataFrame(data)

# Conte as ocorrências de cada categoria
contagem = df['Sexo'].value_counts()

# Calcule os percentuais em relação ao total
percentuais = (contagem / contagem.sum()) * 100

# Crie o gráfico de barras horizontais
plt.barh(contagem.index, contagem.values)

# Adicione rótulos aos eixos
plt.xlabel('Contagem')
plt.ylabel('Sexo')

# Adicione contagem percentual dentro de cada barra
for i, v in enumerate(contagem.values):
    plt.text(v, i, f'{v} ({percentuais[i]:.2f}%)', va='center')

# Exiba o gráfico
plt.show()
```
